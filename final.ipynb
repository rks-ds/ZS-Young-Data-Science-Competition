```python
import math
def ndcg_at_k(r,d, k):
#     r = np.asfarray(r)[:k]
    new_arr=[0]*10
    r=list(np.asfarray(r))
    d=list(np.asfarray(d))
    for i in range(10):
        if r[i]==d[i]:
            new_arr[i]=1
        elif r[i] in d:
            pos=d.index(r[i])
            new_arr[i]=min(((pos+1)/(i+1)),((i+1)/(pos+1)))
        else:
            new_arr[i]=0
    rel=np.asfarray(new_arr)
    result=0
    for i in range(10):
        result+= (pow(2,rel[i])-1)/(math.log((i+2),2))
    return result/4.54355
```
